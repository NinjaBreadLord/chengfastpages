```python
%%html 
<html>
<head>
</head>
<body>
  <form class="search" action="">
    <input class="input" type="text"/>
    <input type="submit"/>
  </form>
  <div id="news-container"></div>
  <table>
    <tr>
      <th>Title</th>
      <th>Description</th>
      <th>Source</th>
    </tr>
  </table>

  <script>
    const apiKey = 'wBn2KgxIdqAZCxawJSL8sWJTShXlc5lU70mmCN5mmqI';
    const url = 'https://api.newscatcherapi.com/v2/latest_headlines?countries=US';

    fetch(url, {
      headers: {
        'X-Api-Key': apiKey
      }
    })
      .then(response => response.json())
      .then(data => {
        const newsContainer = document.getElementById('news-container');

        data.articles.forEach(article => {
          const title = article.title;
          const description = article.summary;
          const source = article.source.name;

          const titleElement = document.createElement('h2');
          titleElement.textContent = title;

          const descriptionElement = document.createElement('p');
          descriptionElement.textContent = description;

          const sourceElement = document.createElement('p');
          sourceElement.textContent = 'Source: ' + source;

          newsContainer.appendChild(titleElement);
          newsContainer.appendChild(descriptionElement);
          newsContainer.appendChild(sourceElement);
        });
      })
      .catch(error => {
        // Handle any errors
        console.error('Error:', error);
      });
  </script>
</body>
</html>
```
